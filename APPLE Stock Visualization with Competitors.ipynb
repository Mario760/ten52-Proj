```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
AAPL = pd.read_csv('Data/AAPL.csv')
AMZN = pd.read_csv('Data/AMZN.csv')
GOOG = pd.read_csv('Data/GOOG.csv')
META = pd.read_csv('Data/META.csv')
NFLX = pd.read_csv('Data/NFLX.csv')
```

```python
AAPL.set_index('Date', inplace=True)
AAPL['Open'].plot(label='APPLE',figsize=(16,8),title='Opening Price')
AMZN['Open'].plot(label='AMAZON')
GOOG['Open'].plot(label='GOOGLE')
META['Open'].plot(label='META')
NFLX['Open'].plot(label='NETFLIX')
plt.legend()
plt.plot()
```

```python
AAPL['High'].plot(label='APPLE',figsize=(16,8),title='High Price')
AMZN['High'].plot(label='AMAZON')
GOOG['High'].plot(label='GOOGLE')
META['High'].plot(label='META')
NFLX['High'].plot(label='NETFLIX')
plt.legend()
plt.plot()
```

```python
AAPL['Close'].plot(label='APPLE',figsize=(16,8),title='Close Price')
AMZN['Close'].plot(label='AMAZON')
GOOG['Close'].plot(label='GOOGLE')
META['Close'].plot(label='META')
NFLX['Close'].plot(label='NETFLIX')
plt.legend()
plt.plot()
```

```python
AAPL['MA50']=AAPL['Open'].rolling(50).mean()
AAPL['MA100']=AAPL['Open'].rolling(100).mean()
AAPL[['Open','MA50','MA100']].plot(label='APPLE',figsize=(16,8));
```
